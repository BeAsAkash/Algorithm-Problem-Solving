```python
import matplotlib.pyplot as plt
import numpy as num
%matplotlib inline
```

```python
Input=[500,1000,2500,5000,10000]
Time=[1,3,14,52,190]
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.set_title('Bubble Sort',fontsize=20)
ax.set_xlabel('<----------INPUT---------->',fontsize=16)
ax.set_ylabel('<----------TIME---------->',fontsize=16)
plt.plot(Input,Time,'-.b',label='curve1')
ax.legend()
```
